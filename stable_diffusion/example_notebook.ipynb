```python

# https://huggingface.co/models?pipeline_tag=text-to-image&p=3&sort=trending

# Packages
import torch
from diffusers import StableDiffusionPipeline

```

```python
# Model to generate with
models = [
    "dreamlike-art/dreamlike-anime-1.0",
    "dreamlike-art/dreamlike-photoreal-2.0",
    "dreamlike-art/dreamlike-diffusion-1.0",
]
# Prompts
prompts = {
    "a":"An onimous apple tree. Impressionist style.",
    "b":"An ominous apple tree. Anime style. Black and white.",
    "c":"A subtley ominous apple tree, looming over a couple. Graphic novel noir style.",
    "d":"A group of cartoon animals in a forest. Adult colouring book style.",
    "e":"A cartoon giraffe from a childrens book. In black and white with only outlines for colouring in.",
    "f":"Outline cartoon giraffe black and white",
    "g":"Outline trees, black and white",
    "h":"perfect day",
    "i":"A stuffed toy giraffe that has come to life, standing in a woods surroundes by friends that are other stuffed toys come to life. Uncluttered and in outline, a colouring book for young adults.",

}

```

```python
i = "b"
model = 0
```

```python
prompt = prompts[i]
model_id = models[model]
```

```python
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
image = pipe(prompt).images[0]
```

```python
display(image)
```

```python
image.save(f'images/result_{model}_{i}.jpg')
```
